**Looking at the data, there are only 50 values for which the data point holds a value of 0 and 450 values for which the data point holds a value of 1. We can maybe provide a more balanced testing and training data**

```python
new_1s = df[df['Y'] == 1][0:50]
new_0s = df[df['Y'] == 0]
new_df = pd.concat([new_1s,new_0s])
new_df = new_df.sample(frac = 1)
new_df
```
